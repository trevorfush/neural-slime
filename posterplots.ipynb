```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm
from palettable.cmocean.sequential import Matter_20_r, Algae_20_r, Thermal_20
from palettable.colorbrewer.sequential import YlOrRd_9_r
from palettable.mycarta import LinearL_3
```

```python
font={
    "size" : 18
}

plt.rc("font", **font)

cmap = Algae_20_r.mpl_colormap
# cmap = Matter_20_r.mpl_colormap
# cmap = Thermal_20.mpl_colormap
# cmap = LinearL_3.mpl_colormap

architecture = "arch2"

for output in tqdm(range(80)):
    output_str = str(output).zfill(4)
    # print(output_str)
    data_elu = np.load(f"standard_elu_{architecture}/STATE{output_str}.npy")
    data_relu = np.load(f"standard_relu_{architecture}/STATE{output_str}.npy")
    data_selu = np.load(f"standard_selu_{architecture}/STATE{output_str}.npy")
    data_exp = np.load(f"standard_exponential_{architecture}/STATE{output_str}.npy")
    data_softplus = np.load(f"standard_softplus_{architecture}/STATE{output_str}.npy")
    data_sigmoid = np.load(f"standard_sigmoid_{architecture}/STATE{output_str}.npy")
    data_tanh = np.load(f"standard_tanh_{architecture}/STATE{output_str}.npy")
    
    # print(data_elu)
    
    fig, ax = plt.subplots(3,3,figsize=(15,15))
    
    ax[0,0].imshow(data_elu, interpolation="spline36", cmap=cmap)
    ax[0,0].axis("off")
    ax[0,0].set_title("elu")
    
    ax[0,1].imshow(data_relu, interpolation="spline36", cmap=cmap)
    ax[0,1].axis("off")
    ax[0,1].set_title("relu")
    
    ax[0,2].imshow(data_selu, interpolation="spline36", cmap=cmap)
    ax[0,2].axis("off")
    ax[0,2].set_title("selu")
    
    ax[1,0].imshow(data_exp, interpolation="spline36", cmap=cmap)
    ax[1,0].axis("off")
    ax[1,0].set_title("exponential")
    
    ax[1,1].imshow(data_softplus, interpolation="spline36", cmap=cmap)
    ax[1,1].axis("off")
    ax[1,1].set_title("softplus")
    
    ax[1,2].imshow(data_sigmoid, interpolation="spline36", cmap=cmap)
    ax[1,2].axis("off")
    ax[1,2].set_title("sigmoid")
    
    ax[2,1].imshow(data_tanh, interpolation="spline36", cmap=cmap)
    ax[2,1].axis("off")
    ax[2,1].set_title("tanh")
    
    ax[2,0].axis("off")
    ax[2,2].axis("off")
    
    fig.tight_layout()
    
    plt.savefig(f"{architecture}_multiplot_{output_str}.png")
    # plt.show()
    plt.close()
```
